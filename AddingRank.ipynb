```python
import ast
import pandas as pd
import numpy as np
import warnings
from winedge import Card, Evaluator
```

```python
# Change to your filepath
pk = pd.read_csv('dataframe/poker_dataframe.csv')

pk_rvr = pk[(~pk['River'].isnull()) & (~pk['SB cards'].isnull()) & (~pk['BB cards'].isnull())]
```

```python
pk_rvr
```

```python
eval = Evaluator()
```

```python
warnings.filterwarnings('ignore')
pk_rvr['SB Handrank'] = np.nan
pk_rvr['BB Handrank'] = np.nan
pk_rvr['SB Hand Strength'] = np.nan
pk_rvr['BB Hand Strength'] = np.nan
```

```python
pk_rvr
```

```python
handSB = ast.literal_eval(pk_rvr['SB cards'][1].replace("[ ", "['").replace(" ]", "']").replace(", ", "', '"))
handSB
```

```python
for loc in pk_rvr.index:
    board_str = ast.literal_eval(pk_rvr['Flop'][loc]) + [pk_rvr['Turn'][loc], pk_rvr['River'][loc]]
    handSB_str = ast.literal_eval(pk_rvr['SB cards'][loc].replace("[ ", "['").replace(" ]", "']").replace(", ", "', '"))
    handBB_str = ast.literal_eval(pk_rvr['BB cards'][loc].replace("[ ", "['").replace(" ]", "']").replace(", ", "', '"))

    handSB = [Card.new(card_str) for card_str in handSB_str]
    handBB = [Card.new(card_str) for card_str in handBB_str]
    board = [Card.new(card_str) for card_str in board_str]

    pk_rvr['SB Handrank'][loc] = eval.class_to_string(eval.get_rank_class(eval.evaluate(handSB, board)))
    pk_rvr['SB Hand Strength'][loc] = eval.get_rank_percentage(eval.evaluate(handSB, board))

    pk_rvr['BB Handrank'][loc] = eval.class_to_string(eval.get_rank_class(eval.evaluate(handBB, board)))
    pk_rvr['BB Hand Strength'][loc] = eval.get_rank_percentage(eval.evaluate(handBB, board))
```

```python
pk_rvr = pk_rvr.reset_index()
```
