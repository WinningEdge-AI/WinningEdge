```python
def check_fold(preflop_actions, flop_actions, turn_actions, river_actions):
    
    fold_sb = False
    fold_bb = False
    
    if preflop_actions is not None:
        for index, action in enumerate(preflop_actions):
            if action[0] == 'f':
                if index % 2 == 0:
                    fold_sb = True
                else:
                    fold_bb = True
    
    if flop_actions is not None and fold_sb == False and fold_bb == False:
        for index, action in enumerate(flop_actions):
            if action[0] == 'f':
                if index % 2 == 0:
                    fold_sb = True
                else:
                    fold_bb = True
    
    if turn_actions is not None and fold_sb == False and fold_bb == False:
        for index, action in enumerate(turn_actions):
            if action[0] == 'f':
                if index % 2 == 0:
                    fold_sb = True
                else:
                    fold_bb = True
    
    if river_actions is not None and fold_sb == False and fold_bb == False:
        for index, action in enumerate(river_actions):
            if action[0] == 'f':
                if index % 2 == 0:
                    fold_sb = True
                else:
                    fold_bb = True
    
    return fold_sb, fold_bb
```

```python
preflop_ex = df['Preflop actions'][1]
flop_ex = df['Flop actions'][1]
turn_ex = df['Turn actions'][1]
river_ex = df['River actions'][1]
```

```python
check_fold(preflop_ex, flop_ex, turn_ex, river_ex)
```

```python
df['fold_sb'], df['fold_bb'] = zip(*df.apply(lambda row: check_fold(row['Preflop actions'], row['Flop actions'], row['Turn actions'], row['River actions']), axis=1))
```
