```python
import ast
import numpy as np
import pandas as pd
import warnings
from winedge import Card, Evaluator
```

```python
# Change to your filepath
pk = pd.read_pickle('dataframe/poker_dataframe.pkl')

pk_rvr = pk[(~pk['River'].isnull()) & (~pk['SB cards'].isnull()) & (~pk['BB cards'].isnull())]
```

```python
eval = Evaluator()
```

```python
warnings.filterwarnings('ignore')
pk_rvr['SB Handrank'] = np.nan
pk_rvr['BB Handrank'] = np.nan
pk_rvr['SB Hand Strength'] = np.nan
pk_rvr['BB Hand Strength'] = np.nan
```

```python
handSB = ast.literal_eval(pk_rvr['SB cards'][1].replace("[ ", "['").replace(" ]", "']").replace(", ", "', '"))
print(handSB)

#board_str = ast.literal_eval(pk_rvr['Flop'][1]) + [pk_rvr['Turn'][1], pk_rvr['River'][1]]
#board_str
turn_river_str = [pk_rvr['Turn'][1],
pk_rvr['River'][1]]
print(turn_river_str)

#flop_str = ast.literal_eval(pk_rvr['Flop'][1].replace("[ ", "['").replace(" ]", "']").replace(", ", "', '"))
#print(flop_str)
flop_list = pk_rvr['Flop'][1]
flop_str = [str(item) for item in flop_list]
print(flop_str)

board_str = [pk_rvr['Turn'][1],
    pk_rvr['River'][1]] + \
    [str(item) for item in flop_list]
print(board_str)
```

```python
for loc in pk_rvr.index:
    board_str = [pk_rvr['Turn'][loc].replace("[ ", "['").replace(" ]", "']").replace(", ", "', '"),
        pk_rvr['River'][loc].replace("[ ", "['").replace(" ]", "']").replace(", ", "', '")] + \
        [str(item) for item in pk_rvr['Flop'][loc]]
    handSB_str = ast.literal_eval(pk_rvr['SB cards'][loc].replace("[ ", "['").replace(" ]", "']").replace(", ", "', '"))
    handBB_str = ast.literal_eval(pk_rvr['BB cards'][loc].replace("[ ", "['").replace(" ]", "']").replace(", ", "', '"))

    handSB = [Card.new(card_str) for card_str in handSB_str]
    handBB = [Card.new(card_str) for card_str in handBB_str]
    board = [Card.new(card_str) for card_str in board_str]

    pk_rvr['SB Handrank'][loc] = eval.class_to_string(eval.get_rank_class(eval.evaluate(handSB, board)))
    pk_rvr['SB Hand Strength'][loc] = eval.get_rank_percentage(eval.evaluate(handSB, board))

    pk_rvr['BB Handrank'][loc] = eval.class_to_string(eval.get_rank_class(eval.evaluate(handBB, board)))
    pk_rvr['BB Hand Strength'][loc] = eval.get_rank_percentage(eval.evaluate(handBB, board))
```

```python

pkeval = Evaluator()
for loc in pk_rvr.index:
        # Extract the board and player cards from the dataframe and format them for use in
        # the Card.py module.
        board_str = \
            [pk_rvr['Turn'][loc].replace("[ ", "['").replace(" ]", "']").replace(", ", "', '"),
            pk_rvr['River'][loc].replace("[ ", "['").replace(" ]", "']").replace(", ", "', '")] + \
            [str(item) for item in pk_rvr['Flop'][loc]]
        handSB_str = ast.literal_eval(pk_rvr['SB cards'][loc].\
                                      replace("[ ", "['").replace(" ]", "']").replace(", ", "', '"))
        handBB_str = ast.literal_eval(pk_rvr['BB cards'][loc].\
                                      replace("[ ", "['").replace(" ]", "']").replace(", ", "', '"))

        # Create unique cards formatted for the evaluator using the Card.py module based on the
        # player and board coards extracted from the dataframe.
        handSB = [Card.new(card_str) for card_str in handSB_str]
        handBB = [Card.new(card_str) for card_str in handBB_str]
        board = [Card.new(card_str) for card_str in board_str]

        pk_rvr['SB Handrank'][loc] = pkeval.class_to_string(
            pkeval.get_rank_class(pkeval.evaluate(handSB, board)))
        pk_rvr['SB Hand Strength'][loc] = pkeval.get_rank_percentage(pkeval.evaluate(handSB, board))

        pk_rvr['BB Handrank'][loc] = pkeval.class_to_string(
            pkeval.get_rank_class(pkeval.evaluate(handBB, board)))
        pk_rvr['BB Hand Strength'][loc] = pkeval.get_rank_percentage(pkeval.evaluate(handBB, board))
```

```python
pk_rvr = pk_rvr.reset_index()
```
