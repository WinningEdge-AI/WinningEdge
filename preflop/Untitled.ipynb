```python
import pandas as pd
```

```python
df = pd.read_csv("../dataframe/poker_dataframe.csv")
filtered_df = df[df['SB cards'].notna()]
```

```python
filtered_df=filtered_df.reset_index(drop=True)
filtered_df
```

```python
def suited_hole_cards(cards):
    """
    This function returns true if hole cards are suited
    """
    if len(cards)!=10:
        raise ValueError("String must contain 10 characters!")
        
    return cards[3]==cards[7]
```

```python
def hole_cards_3_letter_style(cards):
    """
    This function takes the hole cards and translates them to the 3 letter style.
    The output is ordered to show the highest card first
    
    Ex: [ Qh, Tc ]-> QTo because the suits are different
    """
    if len(cards)!=10:
        raise ValueError("String must contain 10 characters!")   
        
    rank_values = {'A': 14, 'K': 13, 'Q': 12, 'J': 11, 'T': 10, '9': 9, '8': 8, '7': 7, '6': 6, '5': 5, '4': 4, '3': 3, '2': 2}
    
    sorted_cards = sorted([cards[2],cards[6]], key=lambda x: rank_values.get(x[0], 0), reverse=True)
    
    if suited_hole_cards(cards):
        return sorted_cards[0] + sorted_cards[1] + 's'
    else:
        return sorted_cards[0] + sorted_cards[1] + 'o'
    
```

```python
filtered_df['SB cards 3style'] = filtered_df['SB cards'].apply(hole_cards_3_letter_style)
filtered_df['BB cards 3style'] = filtered_df['BB cards'].apply(hole_cards_3_letter_style)

# Show the DataFrame with the new column
filtered_df.head()  # .head() just shows the first few rows
```

```python
import pandas as pd

# Sample data: a list of lists with tuples for cell name and value
# Replace this with your actual data
sample_data = [[(f'Name{i}{j}', i*j) for j in range(1, 14)] for i in range(1, 14)]

# Create a pandas DataFrame
df = pd.DataFrame(sample_data)

# Function to display the DataFrame in a 13x13 format with cell names and values
def display_table(data_frame):
    # Creating a displayable table with both name and value
    display_data = pd.DataFrame(index=data_frame.index, columns=data_frame.columns)

    for row in data_frame.iterrows():
        row_index = row[0]
        for col_index in range(len(row[1])):
            cell = row[1][col_index]
            display_data.iloc[row_index, col_index] = f"{cell[0]}: {cell[1]}"

    return display_data

# Display the table
display_table(df)
```
