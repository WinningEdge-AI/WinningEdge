```python
# Eval Flop Straight 
# Risk 0-1 = 3 or more cards required for straight
# Risk 2 = 2 cards required for straight
# Risk 3 = 1 card required for straight
# Risk 4 = straight on board
for loc in st_df.index:
    if st_df['Flop'][loc] is not None:
        rank_char = [card[0] for card in st_df['Flop'][loc]]
        int_rank = set([Card.CHAR_RANK_TO_INT_RANK[char] for char in rank_char])

        if min(int_rank)+4 >= max(int_rank):
            st_df['Flop Straight Risk'][loc] = 2
        else:
            st_df['Flop Straight Risk'][loc] = 1
    else:
        continue
st_df
```

```python
# Turn Eval
for loc in st_df.index:
    if st_df['Turn'][loc] is not None:
        rank_char = [card[0] for card in st_df['Flop'][loc]] + [st_df['Turn'][loc][0]]
        int_rank = sorted([Card.CHAR_RANK_TO_INT_RANK[char] for char in rank_char])

        risk = 0
        for i in range(len(int_rank)-1):
            diff = int_rank[i+1] - int_rank[i]
            if diff == 1 or diff == 2 or diff == 11 or diff == 12:
                risk += 1
            else:
                continue
        if risk < 2:
            risk =1
        st_df['Turn Straight Risk'][loc] = risk
    else:
        continue
st_df
```

```python
# River Eval
for loc in st_df.index:
    if st_df['River'][loc] is not None:
        rank_char = [card[0] for card in st_df['Flop'][loc]] + [st_df['Turn'][loc][0]] + [st_df['River'][loc][0]]
        int_rank = sorted([Card.CHAR_RANK_TO_INT_RANK[char] for char in rank_char])

        risk = 0
        for i in range(len(int_rank)-1):
            diff = int_rank[i+1] - int_rank[i]
            if diff == 1 or diff == 2 or diff == 11 or diff == 12:
                risk += 1
            else:
                continue
        if risk < 2:
            risk =1
        st_df['River Straight Risk'][loc] = risk
    else:
        continue
st_df
```
