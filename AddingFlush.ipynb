```python
import ast
import numpy as np
import pandas as pd
import warnings
from collections import Counter
from winedge import Card, Evaluator
```

```python
pk = pd.read_pickle('dataframe/poker_dataframe_bugfixed.pkl')

pk_rvr2 = pk[(~pk['River'].isnull()) & (~pk['SB cards'].isnull()) & (~pk['BB cards'].isnull())]
```

```python
warnings.filterwarnings('ignore')
pk_rvr2['Board'] = np.nan
pk_rvr2['Flush Count'] = np.nan
pk_rvr2['Flush'] = np.nan
```

```python
pk_rvr2 = pk_rvr2.reset_index()
pk_rvr2
```

```python
for loc in pk_rvr2.index:
    pk_rvr2['Board'][loc] = pk_rvr2['Flop'][loc]
    pk_rvr2['Board'][loc].append(pk_rvr2['Turn'][loc])
    pk_rvr2['Board'][loc].append(pk_rvr2['River'][loc])
    pk_rvr2['Flop'][loc] = pk_rvr2['Flop'][loc][slice(3)]
    
```

```python
pk_rvr2
    
```

```python
for games in pk_rvr2.index:
    suits=[]
    for i in range(5):
        suits.append(pk_rvr2['Board'][games][i][1])
    counts = Counter(suits)
    count = max(counts.values())
    if count == 1:
        Flush = 0
    elif count ==2:
        Flush = 0
    elif count == 3:
        Flush = 1
    elif count == 4:
        Flush = 2
    else:
        Flush = 3
    pk_rvr2['Flush Count'][games] = count
    pk_rvr2['Flush'][games] = Flush
```
